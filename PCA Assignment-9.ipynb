```python
import pandas as pd

# Load the dataset (replace with your file path)
df = pd.read_csv("C:\\Users\\user\\AppData\\Local\\Temp\\a69aba1a-2b24-4a4b-a701-84b78b5d7bf2_PCA.zip.bf2\\PCA\\wine.csv")

# Display basic information about the dataset
print(df.info())
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for each feature
df.hist(figsize=(12, 10), bins=30)
plt.show()

# Box plots for each feature
for column in df.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Box Plot of {column}')
    plt.show()

# Density plots for numerical features
for column in df.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(df[column], shade=True)
    plt.title(f'Density Plot of {column}')
    plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = df.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Fit PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Scree plot
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.grid()
plt.show()

# Determine the optimal number of principal components (e.g., 95% variance)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)
optimal_components = np.where(cumulative_variance >= 0.95)[0][0] + 1
print(f'Optimal number of principal components: {optimal_components}')
```

```python
# Transform the dataset into principal components
pca = PCA(n_components=optimal_components)
X_pca_optimal = pca.fit_transform(X_scaled)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Determine optimal number of clusters using the Elbow method
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.grid()
plt.show()

# Fit K-means with optimal number of clusters
optimal_k = 4  # Replace with the optimal number found
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Evaluate clustering performance
silhouette_avg = silhouette_score(X_scaled, clusters)
print(f'Silhouette Score for original data: {silhouette_avg}')
```

```python
# Scatter plot of the first two principal components (if available)
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters, palette='viridis')
plt.title('Clustering Results (Original Data)')
plt.show()
```

```python
# Fit K-means with PCA-transformed data
kmeans_pca = KMeans(n_clusters=optimal_k, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca_optimal)

# Evaluate clustering performance
silhouette_avg_pca = silhouette_score(X_pca_optimal, clusters_pca)
print(f'Silhouette Score for PCA-transformed data: {silhouette_avg_pca}')
```

```python
# Scatter plot of the first two principal components after PCA
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_pca_optimal[:, 0], y=X_pca_optimal[:, 1], hue=clusters_pca, palette='viridis')
plt.title('Clustering Results (PCA Data)')
plt.show()
```
